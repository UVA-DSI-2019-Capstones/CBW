```python
import os
import numpy as np
import pandas as pd
```

```python
#Set the path where there is CBW_Bess_tags.csv file which is an output from the code CBW - XML Parser.ipynb code
bess_data = pd.read_csv("CBW_Bess_tags.csv")
```

```python
##Shape
print(bess_data.shape)

#Number of NULL values
bess_data.iloc[:,0:3].isnull().sum()
```

```python
bess_data = bess_data[~bess_data.Event.isnull()]
```

```python
##Shape
print(bess_data.shape)

#Number of NULL values
bess_data.iloc[:,0:3].isnull().sum()
```

```python
bess_data.Content[bess_data.Content.isnull()] = ""
```

```python
bess_data['BESS_feature'] = bess_data.iloc[:,[2,1,0]].apply(lambda x: '_'.join(x),axis = 1)
bess_data['values'] = 1
```

```python
print(bess_data.columns)
bess_data.head()
```

```python
bess_data = (bess_data.pivot_table(index=['para no', 'URI', 'author', 'biographyID', 'collectionID','personaName','title'],
                       columns='BESS_feature', values='values', aggfunc=np.sum)).reset_index()
```

```python
bess_data.shape
```

```python
bess_data.to_csv("BESS_transformed.csv")
```
