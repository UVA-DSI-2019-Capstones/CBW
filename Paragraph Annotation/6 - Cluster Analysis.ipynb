```python
import pandas as pd
```

```python
cluters = pd.read_csv('Events_clusters_2.csv')
```

```python
bess_tags = pd.read_csv('CBW_Bess_tags_final2.csv')
bess_tags.head()
```

```python
######### Selecting only the Events ##########
bess_tags  = bess_tags[bess_tags.Type == 'Event']


######### Creating the contents column ########
bess_tags['Content'] = bess_tags['Content'] + ' ; ' + bess_tags['Event']
all_bess = bess_tags['Content'].value_counts()[bess_tags['Content'].value_counts() > 2]
```

```python
cluters.columns = ['Content','Cluster']
```

```python
bess_tags = pd.merge(bess_tags,cluters, on = 'Content')
```

```python
select_cluster = 10
cluster_subset = bess_tags[bess_tags.Cluster == select_cluster]

cluster_subset_agg = cluster_subset.groupby(['collectionID','biographyID','personaName'],as_index = False)['Cluster'].\
                    count().sort_values(['Cluster'],ascending = False)
```

```python
cluster_subset_agg['Percent_cluster'] = 100 * cluster_subset_agg['Cluster']/cluster_subset.shape[0]
```
